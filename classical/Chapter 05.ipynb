```python
# Numerical exploration of exercise 5.2. 

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from tqdm import tqdm
import math

n = 10

k = 1
potential = lambda x, y : 0.5 * k * (x*x + y*y)

tmax, tstep = 10.0, 0.001
t = np.arange(0, tmax, tstep)
tn = t.shape[0]

pos = np.zeros((tn, n, 2))
vel = np.zeros((tn, n, 2))

p = np.concatenate([np.ones((n, 1)), np.ones((n, 1))], axis=1) * math.sqrt(0.5)
v = np.random.randn(n, 2)

for i, time in enumerate(tqdm(t)):
    
    pos[i] = p
    vel[i] = v
    
    f = -k * p
    v += f * tstep
    p += v * tstep
    
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

step = 5
for i in range(n):
    ax.scatter(pos[::step, i, 0], pos[::step, i, 1], s=4, alpha=1.)
    
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
```

```python
# Numerical exploration of exercise 5.2. 

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from tqdm import tqdm

n = 1000

k = .1
potential = lambda x, y : 0.5 * k * (x*x + y*y)

tmax, tstep = 10.0, 0.001
t = np.arange(0, tmax, tstep)
tn = t.shape[0]

pos = np.zeros((tn, n, 2))
vel = np.zeros((tn, n, 2))

p = np.random.randn(n, 2) * 2. - 1.
v = np.random.randn(n, 2)

for i, time in enumerate(tqdm(t)):
    
    pos[i] = p
    vel[i] = v
    
    f = 2 * k * p * 1.0 / (p ** 2).sum(axis=1, keepdims=True)
    v += f * tstep
    p += v * tstep
    
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

step = 5
for i in range(n):
    ax.scatter(pos[::step, i, 0], pos[::step, i, 1], s=4, alpha=0.01)
    
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
```
